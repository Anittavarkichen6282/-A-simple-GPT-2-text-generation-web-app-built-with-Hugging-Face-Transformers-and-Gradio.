```python
import gradio as gr
import torch
from transformers import GPT2LMHeadModel, GPT2Tokenizer
```

```python
tokenizer = GPT2Tokenizer.from_pretrained("gpt2")
model = GPT2LMHeadModel.from_pretrained("gpt2")
model.eval()

# Optional GPU support
device = "cuda" if torch.cuda.is_available() else "cpu"
model.to(device)
```

```python
def generate_text(prompt, max_new_tokens, temperature):
    inputs = tokenizer(prompt, return_tensors="pt").to(device)

    generated_ids = model.generate(
        **inputs,
        max_new_tokens=int(max_new_tokens),
        temperature=float(temperature),
        do_sample=True,
        pad_token_id=tokenizer.eos_token_id
    )

    return tokenizer.decode(
        generated_ids[0],
        skip_special_tokens=True
    )
```

```python
with gr.Blocks() as demo:
    gr.Markdown("# GPT-2 Text Generation")

    with gr.Row():
        input_text = gr.Textbox(
            label="Enter your prompt",
            placeholder="Once upon a time..."
        )
        max_new_tokens_slider = gr.Slider(
            minimum=1, maximum=100, value=80, label="Max New Tokens"
        )
        temperature_slider = gr.Slider(
            minimum=0.1, maximum=1.0, value=0.9, label="Temperature"
        )

    output_text = gr.Textbox(label="Generated Text")
    generate_button = gr.Button("Generate")

    generate_button.click(
        generate_text,
        inputs=[input_text, max_new_tokens_slider, temperature_slider],
        outputs=output_text
    )

demo.launch(share=True, debug=True)
```
